```python
freelancers = {'name':'freelancing Shop','brian': 70, 'black knight':20, 'biccus diccus':100, 'grim reaper':500, 'minstrel':15}
antiques = {'name':'Antique Shop','french castle':400, 'wooden grail':3, 'scythe':150, 'catapult':75, 'german joke':5}
pet_shop = {'name':'Pet Shop','blue parrot':10, 'white rabbit':5, 'newt': 2}

gold = 1000
total_spent = 0
Gold_left = 0
cart= {}
stores =  [freelancers, antiques, pet_shop]
choices = []   

department_store = {}
for store in (freelancers, antiques, pet_shop):
    for item in store:
        if item != 'name':         
            department_store[item] = store[item]
print('Morning inventory of stores',sorted(department_store.items()))
for store in stores:
    for item in store:
        if item != 'name':
            print (item)

    choice = input("What would you like? If you want to skip this store, type 'exit': ")
    choices.append(choice)

    if choice == "exit":
        continue

    elif choice not in store:
        print("Invalid choice!")
        continue

    else:
        price = store[choice]       
        total_spent += price           
        gold -= price  
        cart[store['name']] = choice
        store.pop(choice)

if all(c == "exit" for c in choices):
    print("No cart today, have a great rest of your day!")
else:
    buy_items = ", ".join(cart.values())
    print(f'You Purchased {buy_items}!')
    print(f'That cost you {total_spent}. Thank you for your business!')
    print(f'You have {gold} gold left!')  


department_store_after = {}
for store in (freelancers, antiques, pet_shop):
    for item in store:
        if item != 'name':         
            department_store_after[item] = store[item]
print('evening inventory of stores',sorted(department_store_after.items()))
```
